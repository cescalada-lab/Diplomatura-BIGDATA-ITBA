### Ciclo For
Se repiten las instrucciones del ciclo. Las instrucciones debe estar indentadas. El comando __for__ debe terminar en dos puntos.
El comando __for__ requiere de un elemento iterable (una lista por ejemplo) y una variable que va a ir tomando el valor de cada uno de los elementos del iterable en el orden en que se presentan.

```python
#Puedo utilizar un ciclo for para recorrer la lista e imprimir sus valores junto con una frase.
for nombre in f:
    print('Hola! me presento: soy ', nombre)
```

```python
#Otra forma es recorrerlo con el índice usando la función range. el primer valor debe ser menor que el segundo. Puede haber un tercer valor que es el step.
for i in range(0,len(f),2):
    print('Hola! me presento: soy ', f[i])
```

```python
#Lo puedo recorrer para atrás. El tercer parámetro de range es el STEP. Al indicar -1,
#va descontando en uno el valor del índice. En este caso, el primer valor de range debe ser mayor que el segundo
for i in range(len(f)-1,-1,-1):
    print('Hola! me presento: soy ', f[i])
```

Aplicando la función __enumerate__ al iterable, puedo obtener además del valor de cada elemento, su posición.

```python
for i,j in enumerate(f):
    print(i,j)
```

## Librería Pandas
- Para trabajar con variables multidimensionales más complejas, necesitamos cargar una librería, en este caso __pandas__. A todas las funciones de esta librería las prefijo con _pd_
- __Pandas__ es una de las librerías más usadas en ciencias de datos para realizar preprocesamiento al utilizar herramientas de ML. Aquí se puede ver la <a href="https://pandas.pydata.org/docs/">documentacion</a>

```python
import pandas as pd
```

```python
#Creamos tres variables de tipo lista
edad=[20,30,45,28,36]
sueldo=[80000,84000,200000,25000,30000]
cargo = ["Programador","Analista","Director","Analista jr.","Secretaria"]
```

```python
#Creamos el Data Frame con una columna
personas=pd.DataFrame(edad,columns=["edad"])
personas
```

```python
personas["sueldo"]=sueldo
personas
```

```python
#Agregamos el cargo
personas["cargo"]=cargo
personas
```

```python
#Agregamos una columna con nombre "Mayor18" cuyos valores sean todos True
Mayor18=[True,True,True,True,True]
personas["Mayor18"]=Mayor18
personas
```

```python
#Podemos acceder a una columna completa
personas[['sueldo']]
```

```python
#Podemos acceder a varias columnas 
personas[['sueldo','edad']]
```
